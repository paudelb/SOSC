# 2D Random Walk

```python
def func1(x1,y1):
    '''This function moves one step on randomly selected 
    direction'''
    import random as random
    Axis = random.choice(['x','y'])
    if Axis == 'x':
        a = random.choice([-1,1])
        if a == 1:
            x2 = x1+1
        else:
            x2 = x1-1
            
        y2 = y1
        
    else:
        a = random.choice([-1,1])
        if a == 1:
            y2 = y1+1
        else:
            y2 = y1-1
            
        x2 = x1
    return x2,y2

def func2(N,position):
    '''It will move N steps in 2d'''
    X = []
    Y = []
    '''starting positions'''
    xi=  position[0]
    yi = position[1]
    i = 0
    while i < N:
        x,y = func1(xi,yi)
        X.append(x)
        Y.append(y)
        '''updating position'''
        xi = x
        yi = y
        
        i += 1
    return X,Y


 
 
```

```python
import matplotlib.pyplot as plt
N = 10000
A = func2(N,[0,0])
plt.style.use('seaborn')
plt.figure(figsize = [15,16])
plt.plot(A[0],A[1],"-",color = 'red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Brownian Motion")
plt.show()
```

# 3D Random Walk

```python
def func1(x1,y1,z1):
    '''This function moves one step on randomly selected 
    direction'''
    import random as random
    Axis = random.choice(['x','y','z'])
    if Axis == 'x':
        a = random.choice([-1,1])
        if a == 1:
            x2 = x1+1
        else:
            x2 = x1-1
            
        y2 = y1
        z2 = z1
    if Axis == 'y':
        a = random.choice([-1,1])
        if a == 1:
            y2 = y1+1
        else:
            y2 = y1-1
            
        x2 = x1
        z2 = z1
    

        
    if Axis == 'z':
        a = random.choice([-1,1])
        if a == 1:
            z2 = z1+1
        else:
            z2 = z1-1
            
        x2 = x1
        y2 = y1
    return x2,y2,z2

def func2(N,position):
    '''It will move N steps in 3d'''
    X = []
    Y = []
    Z = []
    '''starting positions'''
    xi=  position[0]
    yi = position[1]
    zi = position[2]
    i = 0
    while i < N:
        x,y,z = func1(xi,yi,zi)
        X.append(x)
        Y.append(y)
        Z.append(z)
        '''updating position'''
        xi = x
        yi = y
        zi = z
        i += 1
    return X,Y,Z


 
 
```

```python
N = 1000
A = func2(N,[0,0,0])

fig = plt.figure(figsize = [15,16])
ax = plt.axes(projection='3d')
zdata = A[0]
xdata = A[1]
ydata = A[2]

plt.style.use('seaborn')
plt.plot(A[0],A[1],A[2],"-",color = 'red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.ylabel("Z-axis")
plt.title("3D Brownian Motion")
plt.show()
```
